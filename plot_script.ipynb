```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix = pd.read_csv("metrices_overview_new.csv", sep="\t")
matrix
```

```python
df = matrix.copy()
# List of column names for each group
group_fpr = ['fet_fpr', 'scores_fpr', 'scoresFet_fpr', 'nbc_fpr', 'nbcFet_fpr', 'n']
#group_recall = ['FET_recall', 'ocsvmR_recall', 'overlappR_recall', 'ocsvmP_recall', 'overlappP_recall', 'NBC_recall']
group_auc = ['fet_auc', 'scores_auc', 'scoresFet_auc', 'nbc_auc', 'nbcFet_auc', 'n']
group_accuracy = ['fet_accuracy', 'scores_accuracy', 'scoresFet_accuracy', 'nbc_accuracy', 'nbcFet_accuracy', 'n']
#group_f1 = ['FET_f1', 'ocsvmR_f1', 'overlappR_f1', 'ocsvmP_f1', 'overlappP_f1', 'NBC_f1']                

# Create separate DataFrames for each group
fpr_df = df[group_fpr]
#recall_df = df[group_recall]
auc_df = df[group_auc]
accuracy_df = df[group_accuracy]
#f1_df = df[group_f1]

fpr_df
```

```python
# FPR

fpr_df_10 = fpr_df[fpr_df['n'] < 11].drop('n', axis='columns')
fpr_df_50 = fpr_df[(fpr_df['n'] < 51) & (fpr_df['n'] > 11)].drop('n', axis='columns')
fpr_df_100 = fpr_df[(fpr_df['n'] < 101) & (fpr_df['n'] > 51)].drop('n', axis='columns')
fpr_df_250 = fpr_df[(fpr_df['n'] < 251) & (fpr_df['n'] > 101)].drop('n', axis='columns')
fpr_df_500 = fpr_df[(fpr_df['n'] < 501) & (fpr_df['n'] > 251)].drop('n', axis='columns')
```

```python
# Define the new x-axis labels
new_labels = ['FET', 'OCSVM', 'OCSVM-FET', 'NBC', 'NBC-FET']

# Calculate mean and standard deviation for each column
mean_values = fpr_df_10.mean()
std_values = fpr_df_10.std()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot with error bars
mean_values.plot(kind='bar', yerr=std_values, capsize=5, ax=ax , color='#81b29a', fontsize = 18)

# Set labels and title
ax.set_ylabel('FPR', fontsize = 18)
#ax.set_xlabel('Approach')
#ax.set_title('False Positive Rate')

# Set new x-axis labels
ax.set_xticklabels(new_labels)
# Set y-axis limits from 0 to 1 in 0.2 steps
ax.set_ylim(0, 0.1)

# Show plot
#plt.show()

# Save the plot as a JPEG file with specified quality (optional)
plt.savefig('fpr_comparison_n10.jpg', dpi=600)
```
